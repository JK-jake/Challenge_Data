### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

```python

print("\nLas categorías con más ventas en la Tienda 1 fueron:")
tn = tienda['Categoría del Producto'].value_counts().head(3)
for i, (tn, cant) in enumerate(tn.items(), 1):
    print(f"  {i}. {tn}: {cant} productos")

print("\nLas categorías con más ventas en la Tienda 2 fueron:")
tn2 = tienda2['Categoría del Producto'].value_counts().head(3)
for i, (tn, cant) in enumerate(tn2.items(), 1):
    print(f"  {i}. {tn}: {cant} productos")

print("\nLas categorías con más ventas en la Tienda 3 fueron:")
tn3 = tienda3['Categoría del Producto'].value_counts().head(3)
for i, (tn, cant) in enumerate(tn3.items(), 1):
    print(f"  {i}. {tn}: {cant} productos")

print("\nLas categorías con más ventas en la Tienda 4 fueron:")
tn4 = tienda4['Categoría del Producto'].value_counts().head(3)
for i, (tn, cant) in enumerate(tn4.items(), 1):
    print(f"  {i}. {tn}: {cant} productos")
```

# 3. Calificación promedio de la tienda


```python
print("\nCalificación promedio de las tiendas")
print("-" * 30)

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], 1):
    rating_promedio = df['Calificación'].mean()
    print(f" Tienda {i}: {rating_promedio:.2f}/5 puntos")
```

# 4. Productos más y menos vendidos

```python
pv = tienda['Producto'].value_counts(), tienda2['Producto'].value_counts(), tienda3['Producto'].value_counts(), tienda4['Producto'].value_counts()

nombres_tiendas = ["tienda", "tienda2", "tienda3", "tienda4"]

for i, conteo in enumerate(pv):
    nombre = nombres_tiendas[i]


    print(f" \\\Análisis de Ventas de la {nombre} //")


    print("\nTop 3 Productos Más Vendidos:")
    print(conteo.head(3))

    print("\nTop 3 Productos Menos Vendidos:")
    print(conteo.tail(3))
```
